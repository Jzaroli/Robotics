```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

# Data import:
heap_k5 = pd.read_csv('./data/twoheap_sim_results_k_5.csv')
heap_k10 = pd.read_csv('./data/twoheap_sim_results_k_10.csv')
heap_k50 = pd.read_csv('./data/twoheap_sim_results_k_50.csv')
heap_k200 = pd.read_csv('./data/twoheap_sim_results_k_200.csv')
heap_k500 = pd.read_csv('./data/twoheap_sim_results_k_500.csv')
sort_k5 = pd.read_csv('./data/quicksort_sim_results_k_5.csv')
sort_k10 = pd.read_csv('./data/quicksort_sim_results_k_10.csv')
sort_k50 = pd.read_csv('./data/quicksort_sim_results_k_50.csv')
sort_k200 = pd.read_csv('./data/quicksort_sim_results_k_200.csv')
sort_k500 = pd.read_csv('./data/quicksort_sim_results_k_500.csv')
```

<h2> Objective </h2>
This analysis takes us through a series of graphs that plot the runtime vs input size for each simulation at each window size: 5, 10, 50, 200, 500. At every window size, both the two heap and the quicksort algorithms were simulated. Their results are compared here. Note: The input sizes were the same for all simulations.  

Let's review the results for the two heap algorithm at a window size of 5 (RunTime is in seconds):

```python
heap_k5
```

```python
plt.scatter(heap_k5['InputSize'], heap_k5['RunTime'], color='blue', label='Two Heap Median Filter', marker='o')
plt.scatter(sort_k5['InputSize'], sort_k5['RunTime'], color='red', label='Quicksort Median Filter', marker='x')
plt.title('Simulation Results with Window Size 5')
plt.xlabel('Input Size (numbers)')
plt.ylabel('Runtime (seconds)')
plt.legend()
plt.show()
```

<h3> Window Size 5 Results </h3>
The quicksort median filter performs better with smaller window sizes.

```python
plt.scatter(heap_k10['InputSize'], heap_k10['RunTime'], color='blue', label='Two Heap Median Filter', marker='o')
plt.scatter(sort_k10['InputSize'], sort_k10['RunTime'], color='red', label='Quicksort Median Filter', marker='x')
plt.title('Simulation Results with Window Size 5')
plt.xlabel('Input Size (numbers)')
plt.ylabel('Runtime (seconds)')
plt.legend()
plt.show()
```

<h3> Window Size 10 Results </h3>
The two algorithms perform similarly at this window size. 

```python
plt.scatter(heap_k50['InputSize'], heap_k50['RunTime'], color='blue', label='Two Heap Median Filter', marker='o')
plt.scatter(sort_k50['InputSize'], sort_k50['RunTime'], color='red', label='Quicksort Median Filter', marker='x')
plt.title('Simulation Results with Window Size 50')
plt.xlabel('Input Size (numbers)')
plt.ylabel('Runtime (seconds)')
plt.legend()
plt.show()
```

<h3> Window Size 50 Results </h3>
The two heap median filter performs better once the window size becomes larger.

```python
plt.scatter(heap_k200['InputSize'], heap_k200['RunTime'], color='blue', label='Two Heap Median Filter', marker='o')
plt.scatter(sort_k200['InputSize'], sort_k200['RunTime'], color='red', label='Quicksort Median Filter', marker='x')
plt.title('Simulation Results with Window Size 200')
plt.xlabel('Input Size (numbers)')
plt.ylabel('Runtime (seconds)')
plt.legend()
plt.show()
```

<h3> Window Size 200 Results </h3>
We see how limited the quicksort algorithm is at large window sizes as it tries to find the median at each step of the process.

```python
plt.scatter(heap_k500['InputSize'], heap_k500['RunTime'], color='blue', label='Two Heap Median Filter', marker='o')
plt.scatter(sort_k500['InputSize'], sort_k500['RunTime'], color='red', label='Quicksort Median Filter', marker='x')
plt.title('Simulation Results with Window Size 200')
plt.xlabel('Input Size (numbers)')
plt.ylabel('Runtime (seconds)')
plt.legend()
plt.show()
```

<h3> Window Size 500 Results </h3>
These results put things further into perspective where for an input size of 128,000 numbers and a window size of 500, the quicksort algorithm takes >5 seconds while the two heap method takes ~ 0.094 seconds

<h2> Conclusion </h2>
The simulations show that the two heap, median filter performs better with medium to large window sizes while the quicksort median filter performs better with small window sizes. This is expected behavior. The two heap method gets a O( n*log(k) ) as opposed to O( n*k*log(k) ) for the quicksort method.
