```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

heap_k5 = pd.read_csv('./data/twoheap_sim_results_k_5.csv')
heap_k50 = pd.read_csv('./data/twoheap_sim_results_k_50.csv')
heap_k200 = pd.read_csv('./data/twoheap_sim_results_k_200.csv')
sort_k5 = pd.read_csv('./data/quicksort_sim_results_k_5.csv')
sort_k50 = pd.read_csv('./data/quicksort_sim_results_k_50.csv')
sort_k200 = pd.read_csv('./data/quicksort_sim_results_k_200.csv')
```

```python
plt.scatter(heap_k5['InputSize'], heap_k5['RunTime'], color='blue', label='Two Heap Filter', marker='o')
plt.scatter(sort_k5['InputSize'], sort_k5['RunTime'], color='red', label='Quicksort Filter', marker='x')
plt.title('Simulation Results with Window Size 5')
plt.xlabel('Input Size (numbers)')
plt.ylabel('Runtime (seconds)')
plt.legend()
plt.show()
```

```python
plt.scatter(heap_k50['InputSize'], heap_k50['RunTime'], color='blue', label='Two Heap Filter', marker='o')
plt.scatter(sort_k50['InputSize'], sort_k50['RunTime'], color='red', label='Quicksort Filter', marker='x')
plt.title('Simulation Results with Window Size 50')
plt.xlabel('Input Size (numbers)')
plt.ylabel('Runtime (seconds)')
plt.legend()
plt.show()
```

```python
plt.scatter(heap_k200['InputSize'], heap_k200['RunTime'], color='blue', label='Two Heap Filter', marker='o')
plt.scatter(sort_k200['InputSize'], sort_k200['RunTime'], color='red', label='Quicksort Filter', marker='x')
plt.title('Simulation Results with Window Size 200')
plt.xlabel('Input Size (numbers)')
plt.ylabel('Runtime (seconds)')
plt.legend()
plt.show()
```

<h2> Conclusion </h2>
The algorithms (in their current form) show that the two heap, median filter performs better with larger window sizes while the quicksort median filter performs better with smaller window sizes.
