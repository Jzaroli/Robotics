<h2> Overview </h2>

The quicksort median filter has a O(n * k * log(k)) while the two heap median filter has a O(n * log(k)); where n is the input size and k is the window size. 
  
Simulations were made to measure the runtime of these two median filter methods. First, to a fixed window size and variable input sizes and then to a fixed input size and variable window sizes.
  
The first batch of simulations ran fixed window sizes and variable input sizes, with each simulation running a window size of 5, 10, 50, 200, and 500 numbers respectively. Then, there is a single simulation looking at a fixed input size of 50,000 numbers and a variable window size, starting at 50 and going up to 450 numbers.

<h2> Analysis with Variable Input Size: </h2>

```python
plt.scatter(heap_k5['InputSize'], heap_k5['RunTime'], color='blue', label='Two Heap Median Filter', marker='o')
plt.scatter(sort_k5['InputSize'], sort_k5['RunTime'], color='red', label='Quicksort Median Filter', marker='x')
plt.title('Simulation Results with Window Size 5')
plt.xlabel('Input Size (numbers)')
plt.ylabel('Runtime (seconds)')
plt.legend()
plt.show()
```

<h3> Window Size 5 Observations </h3>
The quicksort median filter performs better with smaller window sizes. The sorting mechanism is more efficient across the smaller window.

```python
plt.scatter(heap_k10['InputSize'], heap_k10['RunTime'], color='blue', label='Two Heap Median Filter', marker='o')
plt.scatter(sort_k10['InputSize'], sort_k10['RunTime'], color='red', label='Quicksort Median Filter', marker='x')
plt.title('Simulation Results with Window Size 10')
plt.xlabel('Input Size (numbers)')
plt.ylabel('Runtime (seconds)')
plt.legend()
plt.show()
```

<h3> Window Size 10 Observations </h3>
The two algorithms perform similarly at this window size. 

```python
plt.scatter(heap_k50['InputSize'], heap_k50['RunTime'], color='blue', label='Two Heap Median Filter', marker='o')
plt.scatter(sort_k50['InputSize'], sort_k50['RunTime'], color='red', label='Quicksort Median Filter', marker='x')
plt.title('Simulation Results with Window Size 50')
plt.xlabel('Input Size (numbers)')
plt.ylabel('Runtime (seconds)')
plt.legend()
plt.show()
```

<h3> Window Size 50 Observations </h3>
The two heap median filter performs better once the window size becomes larger.

```python
plt.scatter(heap_k200['InputSize'], heap_k200['RunTime'], color='blue', label='Two Heap Median Filter', marker='o')
plt.scatter(sort_k200['InputSize'], sort_k200['RunTime'], color='red', label='Quicksort Median Filter', marker='x')
plt.title('Simulation Results with Window Size 200')
plt.xlabel('Input Size (numbers)')
plt.ylabel('Runtime (seconds)')
plt.legend()
plt.show()
```

<h3> Window Size 200 Observations </h3>
We start to see how limited the quicksort algorithm is with larger window sizes as it tries to find the median within each window.

```python
plt.scatter(heap_k500['InputSize'], heap_k500['RunTime'], color='blue', label='Two Heap Median Filter', marker='o')
plt.scatter(sort_k500['InputSize'], sort_k500['RunTime'], color='red', label='Quicksort Median Filter', marker='x')
plt.title('Simulation Results with Window Size 500')
plt.xlabel('Input Size (numbers)')
plt.ylabel('Runtime (seconds)')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].scatter(heap_k500['InputSize'], heap_k500['RunTime'], color='blue', marker='o')
ax[0].set_title('Two Heap Median Filter Simulation')
ax[0].set_xlabel('Input Size (numbers)')
ax[0].set_ylabel('Runtime (seconds)')

ax[1].scatter(sort_k500['InputSize'], sort_k500['RunTime'], color='red', marker='x')
ax[1].set_title('Quicksort Median Filter Simulation')
ax[1].set_xlabel('Input Size (numbers)')
ax[1].set_ylabel('Runtime (seconds)')

plt.suptitle('Side By Side: Simulation Results with Window Size 500')
plt.show()
```

<h3> Window Size 500 Observations </h3>
These results put things more into perspective where for an input size of 128,000 numbers and a window size of 500, the quicksort algorithm takes > 5 seconds while the two heap method takes ~ 0.094 seconds!

<h2> Analysis with Variable Window Size: </h2>

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].scatter(heap_n50000['WindowSize'], heap_n50000['RunTime'], color='blue', marker='o')
ax[0].set_title('Two Heap Median Filter Simulation')
ax[0].set_xlabel('Window Size (numbers)')
ax[0].set_ylabel('Runtime (seconds)')

ax[1].scatter(sort_n50000['WindowSize'], sort_n50000['RunTime'], color='red', label='Quicksort Median Filter', marker='x')
ax[1].set_title('Quicksort Median Filter Simulation')
ax[1].set_xlabel('Window Size (numbers)')
ax[1].set_ylabel('Runtime (seconds)')

plt.suptitle('Side By Side: Simulation Results with Input Size 50,000')
plt.show()
```

<h3> Input Size 50,000 Observations </h3>
These simulations validate the previous observation that the two heap median filter is vastly more effecient with larger window sizes. 

<h2> Conclusion </h2>
The simulations show that the two heap median filter performs better with larger window sizes while the quicksort median filter performs better with small window sizes. The two heap method gets a O(n * log(k)) as opposed to O(n * k * log(k)) for the quicksort method. The results are in line with expectations and validate the benefits of using a rolling, two heap median filter for mitigating sensor noise in production systems!
